```python
def get_step_graph(feature, bincount): 
    for outcome in animals["OutcomeType"].unique():
        times = animals[feature][animals['OutcomeType']==outcome]
        hist(times, label = outcome, histtype='step', bins=bincount)
    legend(loc='upper left')
```

```python
def age_in_days(age_upon_outcome):
    try:
        days = 0
        str_list = age_upon_outcome.split()
        for i, s in enumerate(str_list):

            if "year" in s:
                days += 365 * int(str_list[i-1])
                # there are lots of "0 year" animals, what do you think we should give them?
            if "month" in s:
                days += 30 * int(str_list[i-1])
            if "week" in s:
                days += 7 * int(str_list[i-1])
            if "day" in s:
                days += int(str_list[i-1])
        return days
    except:
        return np.nan

# adding new feature: "AgeinDays"
animals["AgeinDays"] = animals["AgeuponOutcome"].apply(age_in_days)
```

```python
def is_holiday(t):
    tx_holidays = MyHolidays(state="TX", years =[2015])
    dt = datetime.datetime.strptime(t, r"%Y-%m-%d %H:%M:%S")
    if (dt.month in [1,6,7,8]) or (dt.date() in tx_holidays) or dt.weekday() in [5,6,0]:
        return True
    return False
```
